2) Use the following code to compute the un-normalized posterior pdf (i.e., just the likelihood x prior) on a grid of the linear coefficient (i.e., the slope) of a linear model, with a uniform prior between 0.5 and 1.5. Plot this posterior pdf. Remember this is just a one-dimensional model because the $y$-intercept is fixed. I advise a grid size of 100.

```python
#Functions to do a polynomial fit, and compute the likelihood
def polynomial_fit(theta, x):
    """Polynomial model of degree (len(theta) - 1)"""
    # For a polynomial with order 1, this gives theta_0 + theta_1*x
    # For a polynomial with order 2, this gives theta_0 + theta_1*x + theta_2*x^2, etc.
    return sum(t * x ** n for (n, t) in enumerate(theta))

# compute the data log-likelihood given a model
def logL(theta, data, model=polynomial_fit):
    """Gaussian log-likelihood of the model at theta"""
    x, y, sigma_y = data
    y_fit = model(theta, x)
    return sum(scipy.stats.norm.logpdf(*args) 
               for args in zip(y, y_fit, sigma_y))

# compute the data likelihood given a model
def L(theta, data, model=polynomial_fit):
    """Gaussian log-likelihood of the model at theta"""
    x, y, sigma_y = data
    y_fit = polynomial_fit(theta, x)
    res = scipy.stats.norm.pdf(y, loc=y_fit, scale=sigma_y)
    return np.product(res)
```

```python
#3
def intg(f, a, b, d=0.01, args=None):
    N = int((b-a)/d)
    xx = np.linspace(a, b, N)
    f_arr = f(xx)
    d = xx[1] - xx[0]

    res = (f_arr[0] + f_arr[-1])/2. + sum(f_arr[1:-1])*d

    return res

# x_arr = np.linspace(1e-5,1e5, 10**6)
# plt.loglog(x_arr, post2(x_arr))

%timeit
evidence = intg(postvec, -50, 50)
print('Evidence for linear model is {:.4E}'.format(evidence))
```

```python
#4
def Lvec2D(th1_arr, th2_arr, data=data, model=polynomial_fit):

    #theta0 is fixed
    #th_arr should be a 2D array, row 1 is theta1 and row 2 is theta2
    try:
        L_mat = np.zeros(( len(th1_arr), len(th2_arr) ))
        for i in range(len(th1_arr)):
            for j in range(len(th2_arr)):
                theta = (th0, th1_arr[i], th2_arr[j])
                L_mat[i,j] = L(theta, data, model)
    except:
        theta = (th0, th1_arr, th2_arr)
        L_mat = L(theta, data, model)
    
    return L_mat




# th1_arr = np.linspace(-50,0, 100)
# th2_arr = np.linspace(2.5,2.7, 100)

# xx, yy = np.meshgrid(th1_arr, th2_arr)
# L2d = Lvec2D(th0, th1_arr, th2_arr, data)
# # plt.plot(th1_arr, L2d)
# plt.contourf(xx, yy, L2d)
```

```python
def p1():
    return 1/(1.5-0.5)
def p2():
    return 1/(0.25 + 1)

def post2D(th1_arr, th2_arr):
    try:
        p_mat = np.zeros((th1_arr.size, th2_arr.size))
        for i in range(th1_arr.size):
            for j in range(th2_arr.size):
                theta = ( th0, th1_arr[i], th2_arr[j] )
                p_mat[i,j] = L(theta, data) * p1() * p2()
    except:
        theta = ( th0, th1_arr, th2_arr )
        p_mat = L(theta, data) * p1() * p2()

    return p_mat


th1_arr = np.linspace(-50,0, 100)
th2_arr = np.linspace(2.5,2.7, 100)

xx, yy = np.meshgrid(th1_arr, th2_arr)
p2D = post2D(th1_arr, th2_arr)
plt.contourf(xx, yy, p2D)
```
