1) Use `numpy` to draw $1000$ random samples from a uniform distribution between $1$ and $10$, and store these samples as $x$.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
#Create a random number generator 
rng = np.random.default_rng(seed=1)

#1
x = rng.uniform(1,10,1000)

#2 Histogram of samples
fig, ax = plt.subplots()
ax.hist(x, bins=30, alpha=0.5, density=True)
ax.set_xlabel('Samples from x')
ax.set_ylabel('$p(x)$')
ax.set_title('Probability distribution of random samples drawn from a uniform distribution')
```

```python
#3 Exp of array x
y = np.exp(x)
```

$y = e^{x} \implies x = \log(y)$

The theoretical pdf of y is given by $p(y) = |\frac{dx}{dy}|\, p(x) = e^{-x}\,p(x) = \frac{p(\log y)}{y}$

$p(x)$ is a uniform distribution,

\begin{align}
    p(x) =
    \begin{cases}
    1/9 & 1\,<\,x\,<\,10\\
    0      & \text{otherwise}
    \end{cases}
\end{align}

Therefore, $p(y)$ will be given by

\begin{align}
    p(y) =
    \begin{cases}
    \frac{1}{9\,y} & 1<\log y<10\ ( e < y < e^{10})\\
    0      & \text{otherwise}
    \end{cases}
\end{align}

```python
# 4 Histogram for y
fig, ax = plt.subplots()
ax.hist(y, bins=30, alpha=0.5, density=True)
ax.set_xlabel('Samples from y')
ax.set_ylabel('$p(y)$')
ax.set_title('Probability distribution of transformed distribution')
ax.set_ylim(0., 0.00085)

# Thereotical pdf of y
y_range = np.linspace(np.exp(1), np.exp(10), 100)
py = 1/(9*(y_range))

ax.plot(y_range, py)
```

```python
# 5
#mean of x and y
mean_x = np.mean(x)
mean_y = np.mean(y)
print(f'Mean of x and y: {mean_x}, {mean_y}')

#Median of x and y
med_x = np.median(x)
med_y = np.median(y)
print(f'Median of x and y: {med_x}, {med_y}')
```

```python
#1 Plotting the Gaussian

#Define our pdf as a function
def p(x):
    pi=np.pi
    return np.exp(-x**2/2) / np.sqrt(2*pi) 

x_arr = np.linspace(-3,3,100)
p_arr = p(x_arr)

fig, ax = plt.subplots()
ax.plot(x_arr, p_arr)
ax.set_xlabel('X')
ax.set_ylabel('$p(X)$')
ax.set_title('Gaussian distribution')
```

We can clearly see that the mean is at 0 (where the peak is) and standard deviation is 1 from the general form of a Gaussian, $\frac{1}{\sqrt{2\pi\sigma^2}}\,\exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)$

```python
#3 
N = [10**2, 10**3, 10**4, 10**5]
niter= 100

res2_arr = np.zeros(niter)
res3_arr = np.zeros_like(res2_arr)
res4_arr = np.zeros_like(res2_arr)
res5_arr = np.zeros_like(res2_arr)

for i in range(0,niter):
    
    res2_arr[i] = integrate_gauss_MC(0, 1, N[0], seed=i, nbins=10)
    res3_arr[i] = integrate_gauss_MC(0, 1, N[1], seed=i, nbins=10)
    res4_arr[i] = integrate_gauss_MC(0, 1, N[2], seed=i, nbins=60)
    res5_arr[i] = integrate_gauss_MC(0, 1, N[3], seed=i, nbins=100)
    


#Compilation of results
all_arr = [res2_arr, res3_arr, res4_arr, res5_arr]
mean_arr = [np.mean(arr) for arr in all_arr]
print(mean_arr)
```

```python
#4 Plotting integral value vs logN
logN = np.log10(N)

y1 = [max(arr) for arr in all_arr]
y2 = [min(arr) for arr in all_arr]

fig, ax = plt.subplots()
ax.plot(logN, mean_arr, 'o--')
ax.fill_between(logN, y1, y2, alpha=0.3)
ax.set_xlabel('log N')
ax.set_ylabel('Result of integral')
```

## Problem 3
Let x be a continuous random variable with pdf given by $$p(x)=\frac{1}{2}e^{|x|},$$ for all $x\in\mathbb{R}$. If $y=x^2$, find the cdf of y.

$y = x^2 \implies x = \sqrt(y)$

Let the probability distribution of $y$ be $g(y)$. Then, 

\begin{align}
    g(y)\,dy &= p(x)\,dx \\ 
    g(y) &= \frac{1}{4}\,\frac{e^{-|\sqrt{y}|}}{\sqrt{y}} = \frac{1}{4}\,\frac{e^{-\sqrt{y}}}{\sqrt{y}}
\end{align}

We can ignore the modulus sign since $\sqrt{y}$ is always positive. Note that $y$ is always positive here. 

Let the cdf of y be G(y),

\begin{align}
    G(y) &= \int_0^y g(y')\,dy'\\
         &= \int_0^y \frac{1}{4}\,\frac{e^{-\sqrt{y'}}}{\sqrt{y'}} \,dy'
\end{align}

Substituting $y' = x^2$, 
\begin{align}
    G(y) &= \int_0^y \frac{1}{2}\,e^{-x} \,dx = \frac{1 - e^{-y}}{2}
\end{align}
