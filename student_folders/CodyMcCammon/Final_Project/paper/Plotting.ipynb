```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_csv_data(csv_file_path):
    # Load CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file_path)

    # Specify the columns for x and y axes
    count = "BATTER_SEQ_NUM"
    wOBA_outcome = "EVENT_WOBA_19"

    # Pivot the DataFrame to create a matrix for each unique value in "EVENT_WOBA_19"
    pivot_df = df.pivot_table(index=count, columns=wOBA_outcome, aggfunc='size', fill_value=0)

    # Plot the data as separate lines
    plt.figure(figsize=(10, 6))
    for wOBA_value in [0.0, 0.69, 0.719, 0.87, 1.217, 1.529, 1.94]:
        plt.plot(pivot_df.index, pivot_df[wOBA_value], label=f"Event wOBA: {wOBA_value}")

    # Set plot labels and title
    plt.xlabel("Batter Sequence")
    plt.ylabel("Frequency")
    plt.title(f"Line Plot of Frequency of Event wOBA by Batter Sequence Number")

    # Add legend
    plt.legend()
    plt.grid()
    plt.xlim(0,27)

    # Show the plot
    plt.show()

if __name__ == "__main__":
    csv_file_path = "../TTO_/data/TTO_dataset_510.csv"

    # Call the function to plot the data
    plot_csv_data(csv_file_path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_csv_data(csv_file_path):
    # Load CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file_path)

    # Specify the columns for x and y axes
    count = "BATTER_SEQ_NUM"
    wOBA_outcome = "EVENT_WOBA_19"

    # Pivot the DataFrame to create a matrix for each unique value in "EVENT_WOBA_19"
    pivot_df = df.pivot_table(index=count, columns=wOBA_outcome, aggfunc='size', fill_value=0)

    # Normalize the frequencies
    normalized_df = pivot_df.div(pivot_df.sum(axis=1), axis=0)

    # Plot the data as separate lines
    plt.figure(figsize=(10, 6))
    for wOBA_value in [0.0, 0.69, 0.719, 0.87, 1.217, 1.529, 1.94]:
        plt.plot(normalized_df.index, normalized_df[wOBA_value], label=f"Event wOBA: {wOBA_value}")

    # Set plot labels and title
    plt.xlabel("Batter Sequence")
    plt.ylabel("Normalized Frequency")
    plt.title(f"Normalized Frequency of Event Outcome by Batter Sequence Number")

    # Add legend
    plt.legend()
    plt.xlim(0,27)
    plt.grid()

    # Show the plot
    plt.show()

if __name__ == "__main__":
    csv_file_path = "../TTO_/data/TTO_dataset_510.csv"

    # Call the function to plot the data
    plot_csv_data(csv_file_path)
```
